### ndarray对象

```python
import numpy as np
```

```python
numpy.array(object, dtype = None, copy = True, order = None, subok = False, ndmin = 0)
# object 数组
# dtype 数据类型
# copy 是否需要复制
# order 数组样式，c为反方向，F为列方向，A为任意
# subok 默认返回一个与基类类型一致的数组
# ndmin 指定生成数组的最小维度
```

```python
a = np.array([1,2,3])
print (a)
```

```python
a = np.array([[1,2],[3,4]])
print (a)
```

```python
# 最小维度  
a = np.array([1, 2, 3, 4, 5], ndmin =  2)  
print (a)
```

```python
# dtype 参数  
import numpy as np 
a = np.array([1,  2,  3], dtype = complex)  
print (a)
```

```python
# 使用标量类型
dt = np.dtype(np.int32)
print(dt)
```

```python
# int8, int16, int32, int64 四种数据类型可以使用字符串 'i1', 'i2','i4','i8' 代替
dt = np.dtype('i4')
print(dt)
```

```python
# 将数据类型应用于 ndarray 对象
import numpy as np
dt = np.dtype([('age',np.int8)]) 
a = np.array([(10,),(20,),(30,)], dtype = dt) 
print(a)
```

```python
a = np.arange(24)  
print (a.ndim)             # a 现只有一个维度
# 现在调整其大小
b = a.reshape(2,4,3)  # b 现在拥有三个维度
print (b.ndim)
```

```python
print("a=", a, "\n")
print("b=", b)
```

```python
a = np.array([[1,2,3],[4,5,6]]) 
print (a.shape)
a.shape =  (3,2)  
print (a)
```

```python
# itemsize返回数组每一个元素大小
# 数组的 dtype 为 int8（一个字节）  
x = np.array([1,2,3,4,5], dtype = np.int8)  
print (x.itemsize)
 
# 数组的 dtype 现在为 float64（八个字节） 
y = np.array([1,2,3,4,5], dtype = np.float64)  
print (y.itemsize)
```

### 数据类型

```python
# 创建空数组
x = np.empty([3,2], dtype=int)
print (x)
```

```python
# 创建指定大小数组， 所有元素填充0
np.zeros([3,2], dtype=float, order='C')
```

```python
# 默认为浮点数
x = np.zeros(5) 
print(x)
 
# 设置类型为整数
y = np.zeros((5,), dtype = int) 
print(y)
 
# 自定义类型
z = np.zeros((2,2), dtype = [('x', 'i4'), ('y', 'i4')])  
print(z)
```

```python
# 创建指定形状的数组，数组元素以 1 来填充

# 默认为浮点数
x = np.ones(5) 
print(x)
 
# 自定义类型
x = np.ones([2,2], dtype = int)
print(x)
```

```python
# 创建相同形状的数组
# 填充0
numpy.zeros_like(a, dtype=None, order='K', subok=True, shape=None)
# 填充1
numpy.ones_like(a, dtype=None, order='K', subok=True, shape=None)
```

```python
# 从已有数组创建np数组
# 列表->ndarray
x =  [1,2,3] 
a = np.asarray(x)  
print (a)

# 元组->ndarray
x =  (1,2,3) 
a = np.asarray(x)  
print (a)
```

```python
# np.arrange 从数值范围创建数组

# 生成 0 到 4 长度为 5 的数组:
x = np.arange(5)  
print (x)

# 设置了起始值、终止值及步长：
x = np.arange(10,20,2)  
print (x)
```

```python
# numpy.linspace 函数用于创建一个一维数组，数组是一个等差数列构成的，格式如下：
np.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None)
```

```python
# 设置起始点为 1 ，终止点为 10，数列个数为 10
a = np.linspace(1,10,10)
print(a)
```

```python
a =np.linspace(1,10,10,retstep= True)
 
print(a)
# 拓展例子
b =np.linspace(1,10,10).reshape([10,1])
print(b)
```

```python
# numpy.logspace 函数用于创建一个于等比数列
np.logspace(start, stop, num=50, endpoint=True, base=10.0, dtype=None)
# 底数默认为10
```

```python
a = np.logspace(1.0,  2.0, num =  10)  # a为起始10，终点100，由十个数组成的等比数列
print (a)
```

### 切片和索引

```python
a = np.arange(10)
s = slice(2, 7, 2) # 从索引2开始到7结束，间隔为2
print (a)
print (a[s])
```

```python
b = np.arange(1,10,1).reshape(3,3)
print(b)
print(b[1:])
```

```python
# 显示指定行、列元素
print (b[1, ...]) # 显示第二行
print (b[..., 1]) # 显示第二列
print (b[..., 1:]) # 显示第二列及以后
```
